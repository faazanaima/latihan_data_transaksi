 2. Tambahkan sebuah kolom setelah kolom “Seller Discount” dan beri nama kolom tersebut sebagai
“GMV”. GMV adalah Gross Merchandise Value yang nilainya dihitung berdasarkan “Transaction
Amount” dikurangi “Seller Discount” ditambah dengan “Delivery Fee”.

```python
# Membuat kolom 'GMV' yang berisi data berupa perhitungan value dari pengurangan kolom 'Transaction Amout' dengan kolom 'Seller Discount' yang ditambahkan dengan kolom 'Delivery Fee'
df_transaction['GMV']=df_transaction['Transaction Amount'] - df_transaction['Seller Discount'] + df_transaction['Delivery Fee']
```

```python
# Melakukan reindex untuk memposisikan kolom 'GMV' setelah kolom 'Seller Discount'
df_transaction = df_transaction.reindex(columns=['Paid Date', 'Order Number', 'First Name', 'Last Name','Full Name','Meta Category',
       'Product Name', 'Transaction Amount', 'Seller Discount', 'GMV',
       'Sales Discount', 'Delivery Fee', 'Other Discount'])
df_transaction
```

```python
# Membuat kolom baru 'Group Kategory' berdasarkan 'Meta Category'
group_mapping = {
    'Home/ Food': '1',
    'Sports/ Hobi/ Otomotif': '1',
    'Fashion': '1',
    'Beauty/ Health': '1',
    'Gadget/ Komputer': '2',
    'Elektronik': '2',
    'Service/ Mokado': '2',
    'Babies/ Kids': '2'
}

df_transaction['Group Kategory'] = df_transaction['Meta Category'].map(group_mapping)
```

```python
# Melakukan reindex untuk memposisikan kolom 'Group Kategori' setelah kolom 'Meta Category'
df_transaction = df_transaction.reindex(columns=['Paid Date', 'Order Number', 'First Name', 'Last Name', 'Full Name',
       'Meta Category', 'Group Kategory', 'Product Name', 'Transaction Amount',
       'Seller Discount', 'GMV', 'Sales Discount', 'Delivery Fee',
       'Other Discount'])
df_transaction
```

```python
# Mengubah 'Paid Date' menjadi tipe data datetime
df_transaction['Paid Date'] = pd.to_datetime(df_transaction['Paid Date'], format='%Y%m%d')
print(df_transaction['Paid Date'].dtype)
```

```python
# Menambahkan kolom 'Month_Year' yang berisi format bulan dan tahun
df_transaction['Month_Year'] = df_transaction['Paid Date'].dt.strftime('%Y-%m')
```

```python
# Membuat Pivot Table untuk jumlah total nilai GMV per-bulan untuk setiap group
pv = pd.pivot_table(df_transaction, values='GMV', index=['Month_Year'], columns=['Group Kategory'], aggfunc='sum', fill_value=0)
```

```python
# Membuat DataFrame baru dari Pivot Table
df_gmv_per_month = pd.DataFrame(pv.to_records())
df_gmv_per_month
```

4. Buatlah python statement untuk mencari Seller mana yang memiliki total nilai GMV paling tinggi di bulan Agustus 2017. (Hint: gunakan juga sheet “Seller”)

```python
# Mengubah tipe data pada kolom 'Paid Date' dari int menjadi datetime
df_transaction['Paid Date'] = pd.to_datetime(df_transaction['Paid Date'], format='%Y%m%d')
print(df_transaction['Paid Date'].dtype)
```

```python
# Menggabungkan tabel df_transaction dengan df_seller berdasarkan kolom 'Order Number'
df_merged = pd.merge(df_transaction, df_seller, on="Order Number")
df_merged
```

```python
# Menggabungkan tabel df_transaction dengan df_seller berdasarkan kolom 'Order Number'
df_merged = pd.merge(df_transaction, df_seller, on="Order Number")
df_merged
```

```python
# Membuat filter data dengan ketentuan bulan Agustus 2017 pada kolom 'Paid Date'
filtered_data = df_merged[(df_merged['Paid Date'].dt.month == 8) & (df_merged['Paid Date'].dt.year == 2017)]
```

```python
# Mencari GMV tertinggi per seller dengan grouping filtered_data dengan kunci kolom 'Seller' untuk mengelompokkan 'GMV' yang dijumlahkan.
gmv_per_seller = filtered_data.groupby('Seller')['GMV'].sum().reset_index()
gmv_per_seller
```

```python
# Mencari Seller dengan total GMV tertinggi
seller_highest_gmv = gmv_per_seller.loc[gmv_per_seller['GMV'].idxmax()]
seller_highest_gmv
```

##### Seller yang memiliki total nilai GMV paling tinggi di bulan Agustus 2017 adalah MOBILEPULSA APP dengan GMV 127241500.
